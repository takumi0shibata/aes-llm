```python
import json
import polars as pl

from utils import target_attribute
```

```python
def collect_qwk(model_name):
    results = []

    for prompt_id in range(1, 9):
        result_per_prompt = {}
        for att in target_attribute(prompt_id):
            with open(f"./out/few_shot/{model_name}/{prompt_id}/{att}/summary.json") as f:
                summary_dict = json.load(f)
            result_per_prompt[att] = summary_dict["qwk"]
        results.append(result_per_prompt)

    return pl.DataFrame(results, orient="row")
```

```python
collect_qwk("gemma-4-e4b-it")
```
